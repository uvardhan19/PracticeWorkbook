#### 12. treating the missing values of target column Price:
Here we can see that Price column has more than 20% of missing values.
* Since this column is float type column, we can impute it by mean or median.
* But in this case if we impute such large number(21.71 %) of values in target column, our model will get biased towards mean or median value,which is not a good sign. 
* In such cases if the missing values are around 2-3% then we may impute it.
* Due to these reasons we do not have any option left except removing the rows having price column empty. So we will remove all such rows.

```python
# removing NaN Price rows
df = df[~np.isnan(df['Price'])]
```

#### 13. Checking the percentage of missing values again in each column and treating them

```python
# Percentage of missing values in each column
print(round(100*(df.isnull().sum()/len(df.index)), 2))
```

Here we can see that **Landsize** column has a significant number of missing values i.e. 9.83%. Let us see the description of this column.

```python
df['Landsize'].describe()
```

Here we can see that mean and median of Lansize column are 558.17 and 440 respectively. Also, 75th percentile value is 651 but max value 433014. This represents that data has outliers. Now let us see 90th, 95th and 99th percentile values.

```python
df['Landsize'].describe(percentiles = [0.9, 0.95, 0.99])
```

In such a case ideally we must impute the values either after outlier removal. But since we do not have to remove outliers in this assignment, we are imputing the missing values by median.

```python
df['Landsize'].fillna(df['Landsize'].median(), inplace = True)

# Now lets check the percentage of missing values again
print(round(100*(df.isnull().sum()/len(df.index)), 2))
```

Let us now check the distribution of **Bedroom2, Bathroom, Car, Lattitude, Longtitude** columns

```python
df[['Bedroom2', 'Bathroom', 'Car', 'Lattitude', 'Longtitude']].describe()
```

Here we can see that Bedroom2, Bathroom, Car columns consist of integer values. Lets find out by printing unique values.

```python
df['Bedroom2'].unique()
```

```python
df['Bathroom'].unique()
```

```python
df['Car'].unique()
```

Here we can see that all the unique values in these three columns are integers and therefore it implies that these variables are representing, Number of Bedrroms, Bathrooms or Car. So for all these columns we must impute null values using mode.

```python
df['Bedroom2'] = df['Bedroom2'].fillna(df['Bedroom2'].mode().iloc[0])
df['Bathroom'] = df['Bathroom'].fillna(df['Bathroom'].mode().iloc[0])
df['Car'] = df['Car'].fillna(df['Car'].mode().iloc[0])
```

Now for Lattitude and Longtitude columns we may impute them by mean()

```python
df['Lattitude'] = df['Lattitude'].fillna(df['Lattitude'].mean())
df['Longtitude'] = df['Longtitude'].fillna(df['Longtitude'].mean())
```

In this exercise we got a dataset with a lot of null values and we implemented data cleaning accordingly.
* For the rows having no values at all, we have to remove them.
* For the columns having more than 30% missing values, it will have very less predictive power and we removed it for the sake of simplicity of the model training.
* For rest of the rows we saw if more than 5 columns are missing then the row has very less predictive power. Therefore we removed it.
* For the rest of the columns, we tried to take help of statistical measure and decided the treatment.

As a result we lost 36.29% of data. Although we should avoid loss of data but in case of raw datasets, these kinds of problems may arise. We cannot keep a data which is erroneous or may impart bias when training. In extreme worse cases of data, we should check our data sources also, to get a good quality of data.
